#### Q1. What is the Probability density function?

The Probability Density Function (PDF) is a concept used in probability theory and statistics to describe the likelihood of a continuous random variable taking on a specific value or falling within a particular range of values.

For example, if X represents the height of individuals in a population, the PDF would give us the relative likelihood of finding a person with a particular height. By integrating the PDF over specific height ranges, we can calculate the probability of finding a person within that range.

#### Q2. What are the types of Probability distribution?

There are several types of probability distributions used in probability theory and statistics to model different types of random variables.
1. Discrete Probability Distributions:

- Bernoulli Distribution: Represents a random variable that can take only two outcomes (success or failure) with a fixed probability for success.
- Binomial Distribution: Models the number of successes in a fixed number of independent Bernoulli trials.
- Poisson Distribution: Used to model the number of events that occur in a fixed interval of time or space when events occur independently at a constant rate.
2. Continuous Probability Distributions:

- Uniform Distribution: All outcomes in a range are equally likely.
- Normal Distribution (Gaussian Distribution): Describes many natural phenomena and is characterized by its bell-shaped curve.
- Exponential Distribution: Models the time between events in a Poisson process (events occur continuously and independently at a constant average rate).

#### Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
import math as m

def normal_pdf(x, mean, std):

    exponent = - 1 / 2 * (( x - mean / std ) ** 2 )
    pdf = ( 1 / ( std * m.sqrt( 2 * m.pi ))) * m.exp( exponent )

    return pdf

mean = 0.0
standard_deviation = 1.0
point = 1.5
pdf_at_point = normal_pdf(point, mean, standard_deviation)
print(f"The PDF of the normal distribution at x = {point} is: {pdf_at_point}")
```

#### Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.


The Binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes: success (usually denoted as "S") with probability p and failure (usually denoted as "F") with probability q = 1 - p. The trials are also assumed to be mutually exclusive and independent of each other.
Properties of the Binomial distribution:

- Fixed Number of Trials: The number of trials, denoted as n, is fixed and predetermined.

- Independent Trials: Each trial's outcome does not affect the outcome of any other trial; they are independent events.

- Binary Outcomes: Each trial results in one of two possible outcomes, often referred to as "success" and "failure."

- Constant Probability: The probability of success (p) remains the same for each trial.

- Discrete Distribution: The Binomial distribution is defined for discrete random variables, meaning it applies to a countable set of outcomes.

Formula :

p(x = k) = p^k(1 - p)^(1 - k)

#### Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.4
n = 1000

random_sample = np.random.binomial(n=n, p=p, size=1000)

plt.hist(random_sample, bins=np.arange(n + 1), density=True, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Histogram of Binomial Distribution (n={n}, p={p})')
plt.grid(True)
plt.show()
```

#### Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
import math

def poisson_cdf(x, mean):

    cdf = 0
    for k in range(x + 1):
        p = math.exp(-mean) * math.pow(mean, k) / math.factorial(k)
        cdf += p
    return cdf

x = 2
mean = 3
cdf = poisson_cdf(x, mean)
print(f'CDF at given point {x} is {cdf}')
```

#### Q7. How Binomial distribution different from Poisson distribution?

The Binomial distribution and the Poisson distribution are both probability distributions used to model discrete random variables, but they are different in terms of their assumptions and characteristics. Here are the main differences between the two distributions:

1. Nature of the Process:

- Binomial Distribution: The Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes: success or failure.
- Poisson Distribution: The Poisson distribution models the number of events that occur in a fixed interval of time or space when events occur independently at a constant average rate.
2. Number of Trials or Events:

- Binomial Distribution: The number of trials (n) is fixed and predetermined in the Binomial distribution.
- Poisson Distribution: There is no fixed number of trials or events in the Poisson distribution; it describes a continuous process over time or space.
3. Probability of Success or Event Rate:

- Binomial Distribution: The probability of success in each trial is constant and denoted by p.
- Poisson Distribution: The Poisson distribution does not have a probability of success; instead, it is characterized by a single parameter λ (lambda), which represents the average rate of events per unit of time or space.
4. Outcome Space:

- Binomial Distribution: The Binomial distribution is defined for a discrete outcome space, where the random variable represents the count of successes (non-negative integers).
- Poisson Distribution: The Poisson distribution is also defined for a discrete outcome space, representing the count of events (non-negative integers).

#### Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

lam = 0.4
size = 1000

random_sample = np.random.poisson(lam= lam, size= size)

mean = random_sample.mean()
variance = random_sample.var()

print(f"Mean of random_sample {mean}")
print(f"Variance of random_sample {variance}")
```

#### Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

In both the Binomial distribution and the Poisson distribution, the mean and variance are related to the distribution's parameters. Let's explore the relationship for each distribution:

- Binomial Distribution:
1. Mean (μ): The mean of a Binomial distribution is given by μ = n * p, where n is the number of trials and p is the probability of success in each trial.
2. Variance (σ^2): The variance of a Binomial distribution is given by σ^2 = n * p * (1 - p).
In the Binomial distribution, both the mean and variance depend on the number of trials (n) and the probability of success (p). As n increases, the spread of the distribution becomes broader (larger variance). When p is close to 0.5 and n is large, the Binomial distribution approaches a normal distribution.

- Poisson Distribution:
1. Mean (λ): The mean of a Poisson distribution is equal to its parameter λ, which represents the average rate of events in a fixed interval of time or space.
2. Variance (σ^2): The variance of a Poisson distribution is also equal to its parameter λ.

- For the Binomial distribution, the variance is influenced by both the number of trials and the probability of success. In contrast, for the Poisson distribution, the variance is directly equal to its parameter λ. In both cases, the mean provides a measure of central tendency, while the variance quantifies the spread or dispersion of the distribution around the mean.

#### Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

In a normal distribution, the least frequent data appears in the tails of the distribution, farthest away from the mean. The normal distribution is a symmetric probability distribution, and the mean (μ) is located at the center of the distribution. The tails are the regions in the distribution that extend to the left and right of the mean.

-The normal distribution is characterized by its bell-shaped curve, which means that data near the mean is more common, and the frequency of data decreases as we move away from the mean in either direction. As we move towards the tails of the distribution, data points become increasingly rare.

Specifically:

1. The left tail of the distribution (to the left of the mean) contains data points with values lower than the mean. These data points are less frequent as we move farther to the left.

2. The right tail of the distribution (to the right of the mean) contains data points with values higher than the mean. These data points are less frequent as we move farther to the right.
